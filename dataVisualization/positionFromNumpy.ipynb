```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.load('X.npy')
columns = []
for i in range(33*4):
    columns.append('pose' +str(i))
for i in range(21*3):
    columns.append('lh' +str(i))
for i in range(21*3):
    columns.append('rh' +str(i))

frame = X[:, 0, :]
print(frame.shape)
```

```python
df = pd.DataFrame(frame,  columns=columns)
feature = 'rh8'
plt.figure(1)
plt.subplot(131)
sns.distplot(df[feature])


plt.subplot(132)
df[feature].plot.box()

plt.subplot(133)
plt.scatter(range(1, df[feature].shape[0] + 1), df[feature])
plt.show()
```
